```python
import os
import csv
from PIL import Image
import PIL
import matplotlib.pyplot as plt
import pickle
```

```python
styles = ["Romanticism", "Impressionism", "Realism", "Post-Impressionism"]
```

```python
with open('../landscapes/train_info.csv',  encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    labels = []
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            if row[4] == 'landscape' and row[3] in styles:
                print(row[0])
                source = "./train/train/" + row[0]
                destination = "./landscapes/" + row[0]
                save_center_crop(source, destination)
                labels.append([row[0],styles.index(row[3]),row[3]])
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
def save_center_crop(path, destination):
    im = Image.open(open(path, 'rb'),'r')
    width, height = im.size   # Get dimensions
    
    dim = min(width, height)

    left = (width - dim)/2
    top = (height - dim)/2
    right = (width + dim)/2
    bottom = (height + dim)/2
    
#     plt.imshow(im)
#     plt.show()
    # Crop the center of the image
    im = im.crop((left, top, right, bottom))
#     plt.imshow(im)
#     plt.show()
    
    #resize to 512
    im = im.resize((512,512))
#     plt.imshow(im)
#     plt.show()
    
    im.save(destination)
```

```python
with open('labels.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    for label in labels:
        writer.writerow(label)
```
